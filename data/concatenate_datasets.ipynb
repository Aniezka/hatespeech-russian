```python
import pandas as pd
```

```python
df = pd.read_csv("2_preprocessed_data.csv", low_memory=False)
df.info()
```

```python
data = pd.read_csv('toxicity_classifier_output_sorted.csv')
data.head()
```

```python
len(data)
```

```python
data_merged = data.merge(df, on=["Реплика 1", "Реплика 2"], how='left')
```

```python
data_merged.head()
```

```python
cols = list(data.columns)
cols.remove('Unnamed: 0')

data_unique = data_merged.drop_duplicates(subset=cols + ["Unnamed: 0_x"])
```

```python
del data_unique['Unnamed: 0_x']
```

```python
del data_unique['Unnamed: 0_y']
```

```python
data_unique.head()
```

```python
data_unique.rename(columns={'Реплика 1': 'Реплика 1 toxicity', 'Реплика 2': 'Реплика 2 toxicity'}, inplace=True)
```

```python
data_unique.head()
```

```python
data_unique = data_unique[(data_unique['Реплика 1 toxicity'].apply(lambda x: 25<len(str(x))<500)
           & (data_unique['Реплика 2 toxicity'].apply(lambda x: 25<len(str(x))<500)))]
```

```python
data_unique.info()
```

```python
data_unique.to_csv("3_full_toxicity_output.csv", index=False)
```
