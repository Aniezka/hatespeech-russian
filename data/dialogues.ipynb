# 1 –≤—ã–≥—Ä—É–∑–∫–∞, original

```python
import requests
import json
from types import SimpleNamespace
import ast
import numpy as np
```

```python
url = "https://api.aicloud.sbercloud.ru/public/v2/boltalka/predict"
```

```python
a = [1, 2]
a[0:2]
```

```python
our_file = []
batch = 200
for i in range(0, len(data_new), batch):
    try:
        sl = data_new[i:min(i+batch, len(data_new))]
        dict_1 = {"instances": [{"contexts": sl}]}
        r = requests.post(url,json = dict_1)
        x = json.loads(r.text, object_hook=lambda d: SimpleNamespace(**d))
        y = ast.literal_eval(x.responses)
        y = [n.strip() for n in y]
        for e in y:
            our_file.append(e)
        print(len(our_file) / len(data_new) * 100)
    except Exception:
        our_file.append([])
our_file = np.array(our_file)
pd.DataFrame(our_file).to_csv("replika_2_toxicity_1.csv")
```

```python
tmp_latest_ds = latest_ds.copy()

#—É–¥–∞–ª—è–µ–º 2939
delete_for_each = {
    '–¥–æ–ª–±–æ–µ–±': 730,
    '–¥–µ–±–∏–ª': 370,
    '–ø–∏–¥–æ—Ä–∞—Å': 260,
    '–¥–æ–ª–±–∞–µ–±': 210,
    '–ø–∏–¥–æ—Ä': 170,
    '–∂–∏–¥': 150,
    '—Ö—É–µ—Å–æ—Å': 130,
    '–µ–±–∞–Ω–∞—Ç': 110, 
    '–µ–±–ª–∞–Ω': 100,
    '–∫—É–∫–æ–ª–¥': 80,
    '–ø–∏–∑–¥–∞–±–æ–ª': 70,
    '–∏–¥–∏–æ—Ç': 50,
    '—Ö–µ—Ä': 40,
    '—Ö—É–π–ª–æ': 40, 
    '—É–µ–±–æ–∫': 20,
    '–∞–ª–∫–∞—à': 10,
    '–±–ª—è–¥—å': 153,
    '—Å—É–∫–∞': 125,
    '—à–ª—é—Ö–∞': 121
}

for word, num in delete_for_each.items():
    arr = []
    balanced = []
    
    for ind, row in tmp_latest_ds.iterrows():
        if word == row["–í—ã–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–ª–æ–≤–∞"]:
            arr.append(row)
        else:
            balanced.append(row)

    latest_ds_word = pd.DataFrame(arr)
    latest_ds_word = latest_ds_word[:len(latest_ds_word) - num]

    tmp_latest_ds = pd.concat([pd.DataFrame(balanced), latest_ds_word])
```
