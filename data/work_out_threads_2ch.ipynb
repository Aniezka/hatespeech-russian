```python
import os
import pandas as pd
import re
from tqdm import tqdm
from time import sleep
```

```python
post_id_answer_id = pd.DataFrame(columns=["post_id", "answer_id"])
post_id_post_text = pd.DataFrame(columns=["post_id", "post_text"])
```

```python
import glob, os

os.chdir("../threads/")
for file in glob.glob("*.csv"):
    if "post_id_answer" in file:
        post_id_answer_id = pd.concat([post_id_answer_id, pd.read_csv(f"../threads/{file}")])
    if "post_id_text" in file:
        post_id_post_text = pd.concat([post_id_post_text, pd.read_csv(f"../threads/{file}")])
```

```python
post_id_post_text_dict = {k: v for k, v in post_id_post_text.values}
```

```python
def replace_post_id_by_text(id_: int):
    """
    Return text by post_id
    """
    return post_id_post_text_dict[id_] if id_ in post_id_post_text_dict.keys() else ""
```

```python
post_id_answer_id["post_id_text"] = post_id_answer_id["post_id"].apply(replace_post_id_by_text)
post_id_answer_id["answer_id_text"] = post_id_answer_id["answer_id"].apply(replace_post_id_by_text)
```

```python
def preprocess(text: str):
    """
    Preprocess
    """
    text = str(text)
    filt_url = re.compile(r'https?:\/\/.*[\r\n]*')
    filt_tags = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    text = re.sub(filt_url, ' ', text)
    text = re.sub(filt_tags, ' ', text)
    text = re.sub('([А-Я]{1}[а-я]+)', r' \1', text)
    text = text.replace(">", " ")
    text = text.replace("  ", " ")
    return text
```

```python
post_id_answer_id["post_id_text"] = post_id_answer_id["post_id_text"].apply(preprocess)
post_id_answer_id["answer_id_text"] = post_id_answer_id["answer_id_text"].apply(preprocess)
```

```python
post_id_answer_id = post_id_answer_id[post_id_answer_id["post_id_text"].str.len() > 25]
post_id_answer_id = post_id_answer_id[post_id_answer_id["answer_id_text"].str.len() > 25]
```
