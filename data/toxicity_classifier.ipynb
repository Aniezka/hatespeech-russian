```python
import os,re
import torch
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup 
from tqdm import tqdm

from transformers import BertTokenizer, BertForSequenceClassification
```

```python
# load tokenizer and model weights
tokenizer = BertTokenizer.from_pretrained('SkolkovoInstitute/russian_toxicity_classifier')
model = BertForSequenceClassification.from_pretrained('SkolkovoInstitute/russian_toxicity_classifier')
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
def decide(text):
    res = parse_text(text)[0][0]
    # softmax(res[0][i].detach().numpy())
    return bool(res[0] < res[1]), softmax(res.detach().numpy())
```

```python
def parse_text(text, model=model, tokenizer=tokenizer):
    # prepare the input
    batch = tokenizer.encode(text, return_tensors='pt')
    # inference
    return model(batch)
```

```python
def decide_batches(texts, batch_size):
    res = parse_batches(texts, batch_size)
    answers, scores = [], []
    for i in range(len(res)):
        answers+=[bool(res[0][i][0] < res[0][i][1]) for i in range(len(res[i]))]
        scores += [softmax(res[0][i].detach().numpy()) for i in range(len(res[i]))]
    return answers, scores
```

```python
def parse_batches(texts, batch_size, model=model, tokenizer=tokenizer):
    results = []
    for i in range(0, len(texts), batch_size):
        try:
            #print(i)
            batch = tokenizer(texts[i:i + batch_size], return_tensors='pt', padding=True)
            results += model(**batch)#['logits'].argmax(1).float().data.tolist()
        except:
            pass
    return results
```

```python
def clear(text):
    if '<stage>' in text:
        text = re.split('<stage>', text)[0]
    text = BeautifulSoup(text, "lxml")
    text = text.get_text() 
    text = re.sub(r'\s+', ' ', text)
    return text
```

```python
def process_file(xml):
    one = open(os.path.join(wdir, xml), 'r').read()
    replicas = re.split('<p>',re.split('<body>', one)[1])
    replicas = [clear(re.split('</p>',r)[0]) for r in replicas]
    #checking toxicity
    toxic = [r for r in replicas if decide(r)]
    return toxic
```

```python
replika_1 = [] 
replika_2 = []
res_1 = []
res_2 = []
sc_1 = []
sc_2 = []
 
for i in tqdm(range(len(texts))):
 
    #checking toxicity 
    results1, scores1 = decide(texts[i])
    results2, scores2 = decide(texts[i+1])
 
    if (scores1[1] > 0.97) or (scores2[1] > 0.97 ): 
        replika_1.append(texts[i]) 
        replika_2.append(texts[i+1])
        res_1.append(results1)
        res_2.append(results2)
        sc_1.append(scores1)
        sc_2.append(scores2)
```
