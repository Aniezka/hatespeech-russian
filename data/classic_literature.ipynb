```python
import urllib.request 
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Koziev/NLP_Datasets/master/Conversations/Data/ru.conversations.txt"
file = urllib.request.urlopen(url)
```

```python
full_text = ''
for line in file:
    text = line.decode("utf-8")
    full_text += text
```

```python
len(full_text.split('\n- '))
```

```python
evens = []
odds = []
for i,text in enumerate(full_text.split('\n- ')):
    if (i % 2) == 0:
        evens.append(text)
    else:
        odds.append(text)
```

```python
data = pd.DataFrame(columns = ['Реплика 1', 'Реплика 2'])
for i in range(len(evens)):
    data = data.append({'Реплика 1': evens[i], 'Реплика 2': odds[i]}, ignore_index = True)
```
