```python
import pandas as pd
from razdel import tokenize, sentenize
from tqdm.auto import tqdm
import re
from tqdm.auto import tqdm
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
df = pd.read_csv('синонимы к таргетным группам/result_filter_95.csv')
```

```python
def lemm_text_ru(text):
    
    tokens = tokenize(text)
    words = [_.text for _ in tokens]
    # print(words)
    
    # оставляем только текст
    new_words = []
    for word in words:
    
    # проверяем на стоп-слова и убираем их
          wr = morph.parse(str(word))[0].normal_form
          new_words.append(wr)

    return new_words
```

```python
df['split_rep1'] = df['Реплика 1'].apply(lemm_text_ru)
```

```python
df['split_rep2'] = df['Реплика 2'].apply(lemm_text_ru)
```

```python
new_list_eb = []
new_list_blad = []
new_list_zad = []
new_list_maz = []
new_list_ish = []
new_list_mig = []
new_list_child = []
```

```python
def find_words(text):
  

  for word in text:

    eb = re.search(r'(.*еб)$', word)
    if eb:
      new_word1 = eb.group(1)
      if len(new_word1) > 4:
        new_list_eb.append(new_word1)

    blad = re.search(r'(.*бля[т|д]ь)$', word)
    if blad:
      new_word2 = blad.group(1)
      if len(new_word2) > 6:
        new_list_blad.append(new_word2)

    zad = re.search(r'(.*задый)$', word)
    if zad:
      new_word3 = zad.group(1)
      if len(new_word3) > 6:
        new_list_zad.append(new_word3)

    jop = re.search(r'(.*жопый)$', word)
    if jop:
      new_word4 = jop.group(1)
      if len(new_word4) > 6:
        new_list_zad.append(new_word4)

    maz = re.search(r'(.*мазый)$', word)
    if maz:
      new_word5 = maz.group(1)
      if len(new_word5) > 6:
        new_list_maz.append(new_word5)


    ish = re.search(r'(.*ыш)$', word)
    if ish:
      new_word6 = ish.group(1)
      if len(new_word6) > 4:
        new_list_ish.append(new_word6)

    nig = re.search(r'^(ниг.*)$', word)
    if nig:
      new_word7 = nig.group(1)
      if len(new_word7) > 4:
        new_list_mig.append(new_word7)
    
    hac = re.search(r'^(хач.*)$', word)
    if hac:
      new_word8 = hac.group(1)
      if len(new_word8) > 4:
        new_list_mig.append(new_word8)

    gryz = re.search(r'(.*грыз)$', word)
    if gryz:
      new_word9 = gryz.group(1)
      if len(new_word9) > 6:
        new_list_child.append(new_word9)

    shkol = re.search(r'(шк[о|и]л.*[о|к|а])$', word)
    if shkol:
      new_word10 = shkol.group(1)
      if len(new_word10) > 6:
        new_list_child.append(new_word10)
```

```python
for i, row in tqdm(df.iterrows()):
  find_words(row.split_rep1)
  find_words(row.split_rep2)
```
