```python
contexts_2 = {}

for nec_id in tqdm(get_ids):
  good_labels_dict = defaultdict(list)

  new_df = df[df['context_id'] == nec_id]

  for ind, row in new_df.iterrows():
    
    if row['context_id'] == nec_id:

      label = row['label']

      try:
        good_labels_dict[label].append(float(row['confidence']))
      except:
         good_labels_dict[label].append(np.nan)
      # print(nec_id, row['context_id'], '\n----------')
  
  # print(good_labels_dict)

  list_for_params = []
  if 'good' in good_labels_dict:
    # print(sorted(good_labels_dict['good'], reverse=True))
    list_for_params = ['good', sorted(good_labels_dict['good'], reverse=True)[0]]
  if 'good' not in good_labels_dict:
    
    if 'neutral' in good_labels_dict:
      # print('neutral')
      list_for_params = ['neutral', sorted(good_labels_dict['neutral'], reverse=True)[0]]
    if 'neutral' not in good_labels_dict:
      # print('bad')
      # в любом случае уверенность в выборе должна быть высокой
      # print(sorted(good_labels_dict['bad'], reverse=True))
      if len(good_labels_dict['bad'])==0:
        list_for_params = ['bad', '']
      else:
        list_for_params = ['bad', sorted(good_labels_dict['bad'], reverse=True)[0]]
  # print(list_for_params)
  # params_contexts_2.append(list_for_params)

  cur_string = new_df[(new_df['label'] == list_for_params[0]) & (new_df['confidence'] == str(list_for_params[1]))]
  # print(cur_string)
  # if row['context_id']

  try:
    contexts_2[nec_id] = {'Реплика 1': cur_string['context_0'].tolist()[0], 'Реплика 2': cur_string['reply'].tolist()[0]}
  except:
    continue
```
